```python
try:
    file1 = open ("arpan.txt","r")
except:
    print("file can not be opened")

for line in file1:
    line = line.rstrip()
    if re.search("^hello", line):#only find that line which start with the word "hello"
        print(line)
file1.close()
```

```python
try:
    file1 = open ("arpan.txt","r")
except:
    print("file can not be opened")

for line in file1:
    line = line.rstrip()
    if re.search("you$", line):#only find that line which end with the word "you"
        print(line)
file1.close()
```

## write a program to extract lines starting with the word "From"(or "from")
#and ending with "edu"


```python
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
count=0
for line in file1:
    count+=1
    line = line.rstrip()
    pattern = "^[From from].*edu$"
    if re.search(pattern,line):#line startwith fromand some where u will be there(+ one ore more preceding char)
        print("Line number:",count, "Line: ",line)
print(count)
file1.close()
```

```python
#extract all the lines ending with any digit
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
    
for line in file1:
    line = line.rstrip()
    #pattern = "\d$"
    pattern = "[0-9]$"
    if re.search(pattern,line):
        print(line)

file1.close()
```

```python
#other than lowercase alphabet and digit
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
count=0    
for line in file1:
    line = line.rstrip()
    pattern = "^[^0-9a-z]"
    if re.search(pattern,line):
        count=count+1
        print(line)
print(count)
file1.close()
```

```python
#start with capital folowd by zero or more char but must be end with any didgit
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
    
for line in file1:
    line = line.rstrip()

    pattern = "^[A-Z].*[0-9]$"
    if re.search(pattern,line):
        print(line)

file1.close()
```

```python
#Fri, 04 Jan 2008 18:10:48 -0500
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
    
list3 = []    
for line in file1:
    line = line.rstrip()
    line = re.findall("^[a-zA-Z]{3}.* ([0-9][0-9]):",line)
    if len(line)>0:
        list3.extend(line)
print(list3)
```

```python
try:
    file1 = open ("mbox-short.txt","r")
except:
    print("file can not be opened")
list2 = []   
list1 = []    
for line in file1:
    line = line.rstrip()
    reg1 = re.findall("^[a-zA-Z]{3}.* ([0-9][0-9]):",line)
    reg2 = re.findall("^From.* ([0-9][0-9]):",line)
    if len(reg1)>0:
        list1.extend(reg1)
    if len(reg2)>0:
        list2.extend(reg2)
        
list3 = list1+list2

freq =[]
for i in list3:
    freq.append((i,list3.count(i)))
hist = list(set(freq))
hist.sort(key=lambda x:(x[1],x[0]),reverse=True)
#print(list1)
#print(list2)
print(hist)
```
